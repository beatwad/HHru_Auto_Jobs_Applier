# Login

```python
from typing import List

import time
import random
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys

def pause(low: int, high: int) -> None:
    """
    Выдержать случайную паузу в диапазоне от 
    low секунд до high секунд
    """
    pause = round(random.uniform(low, high), 1)
    time.sleep(pause)

service = Service(executable_path=ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

# Зайти на сайт
driver.get("https://hh.ru")

# Найти кнопку входа
driver.find_element("css selector", '[data-qa="login"]').click()
assert driver.current_url.startswith("https://hh.ru/account/login"), "Wrong login page URL address"
assert driver.title ==  "Вход в личный кабинет", "Wrong login page title"
pause(0, 2)

# Записать логин в поле логина
email = "beatwad@gmail.com"
driver.find_element("name", 'login').send_keys(email)
driver.find_element("css selector", '[data-qa="account-signup-submit"]').click()
pause(0, 2)

# Дождаться, пока пользователь авторизуется в системе
profession = "Data Scientist"

while True:
    try:
        driver.find_element("css selector", '[data-qa="search-input"]').send_keys(profession)
    except NoSuchElementException:
        pause(10, 10)
        print("Пожалуйста, войдите в аккаунт HH ")
        continue
    else:
        break

# driver.find_element("css selector", '[data-qa="search-button"]').click()
# pause(0, 3)
```

# Advanced search

```python
def select_suggestion(driver: webdriver.Chrome) -> None:
    """
    Функция для поиска и выбора первой подсказки в списке подсказок, 
    появляющихся во время поиска
    """
    suggestion_list = driver.find_elements("class name", "suggest__item")
    if len(suggestion_list) > 0:
        suggestion_list[0].click()


def find_by_text_and_click(text: str, driver: webdriver.Chrome) -> None:
    """Функция для поиска элемента по тексту и клика по нему"""
    element = driver.find_element("xpath", f'//*[text()="{text}"]')
    driver.execute_script("arguments[0].scrollIntoView();", element)
    element.click()


def find_by_data_qa_and_click(text: str, driver: webdriver.Chrome) -> None:
    """Функция для поиска элемента по тексту и клика по нему"""
    element = driver.find_element("css selector", f'[data-qa="{text}"]')
    driver.execute_script("arguments[0].scrollIntoView();", element)
    element.click()


def select_checkbox(options: List[str], driver: webdriver.Chrome, type="text") -> None:
    """Функция для выбора всех нужных опций в checkbox"""
    for option in options:
        if type == "text":
            find_by_text_and_click(option, driver)
        else:
            find_by_data_qa_and_click(option, driver)


assert driver.current_url.startswith("https://hh.ru/search/vacancy/advanced")

# Ключевые слова
keywords = "Data scientist"
driver.find_element("css selector", '[data-qa="vacancysearch__keywords-input"]').send_keys(keywords)
pause(1, 2)
select_suggestion(driver)
pause(0, 2)

# Искать только
search_only = ["в описании вакансии"]
for s_o in search_only:
    assert s_o in ["в названии вакансии", "в названии компании", "в описании вакансии"]
    find_by_text_and_click(s_o, driver)


# Исключить слова
words_to_exclude = "Analyst"
driver.find_element("css selector", '[data-qa="vacancysearch__keywords-excluded-input"]').send_keys(words_to_exclude)
pause(1, 2)


# Cпециализация
# Выбрать меню 'Указать специализации'
specialization = "Программист"
find_by_text_and_click("Указать специализации", driver)
pause(1, 2)

# Ввести специализацию в поисковую строку
driver.find_element("css selector", '[data-qa="bloko-tree-selector-popup-search"]').send_keys(specialization)
pause(1, 2)

# Попытаться найти специализацию с таким же названием, что и название специализации в переменной
specialization_list = driver.find_elements("xpath", f'//*[text()="{specialization}"]')
# Иначе выбрать первую специализацию в чекбоксе (если есть, иначе закрыть)
if len(specialization_list) == 0:
    specialization_list = driver.find_elements("css selector", '[data-qa^="bloko-tree-selector-item-text bloko-tree-selector-item-text"]')
pause(1, 2)

if len(specialization_list) > 0:
    driver.execute_script("arguments[0].scrollIntoView();", specialization_list[0])
    specialization_list[0].click()
    driver.find_element("css selector", '[data-qa="bloko-tree-selector-popup-submit"]').click()
else:
    driver.find_element("css selector", '[data-qa="bloko-modal-close"]').click()
pause(1, 2)


# Отрасль компании
# Выбрать меню 'Указать отрасль компании'
industry = "Банк"
find_by_text_and_click("Указать отрасль компании", driver)
pause(1, 2)

# Ввести отрасль в поисковую строку
driver.find_element("css selector", '[data-qa="bloko-tree-selector-popup-search"]').send_keys(industry)
pause(1, 2)

# Попытаться найти отрасль с таким же названием, что и название отрасли в переменной
industry_list = driver.find_elements("xpath", f'//*[text()="{industry}"]')
# Иначе выбрать первую отрасль в чекбоксе (если есть, иначе закрыть)
if len(industry_list) == 0:
    industry_list = driver.find_elements("css selector", '[data-qa^="bloko-tree-selector-item-text bloko-tree-selector-item-text"]')
pause(1, 2)

if len(industry_list) > 0:
    driver.execute_script("arguments[0].scrollIntoView();", industry_list[0])
    industry_list[0].click()
    driver.find_element("css selector", '[data-qa="bloko-tree-selector-popup-submit"]').click()
else:
    driver.find_element("css selector", '[data-qa="bloko-modal-close"]').click()
pause(1, 2)


# Регион
region = "Москва"
driver.find_element("css selector", '[data-qa="advanced-search-region-add"]').send_keys(region)
pause(1, 2)
select_suggestion(driver)
pause(1, 2)


# Уровень дохода
desired_income = 300000
driver.find_element("css selector", '[data-qa="advanced-search-salary"]').send_keys(desired_income)


# Образование
education = ["Высшее"]
for e in education:
    assert e in ["Не требуется или не указано", "Среднее профессиональное", "Высшее"]
select_checkbox(education, driver)


# Требуемый опыт работы
experience = "От 1 года до 3 лет"
assert experience in ["Не имеет значения", "Нет опыта", "От 1 года до 3 лет", "От 3 до 6 лет", "Более 6 лет"]
find_by_text_and_click(experience, driver)


# Тип занятости
employment = ["Полная занятость"]
for e in employment:
    assert e in ["Полная занятость", "Частичная занятость", "Проектная работа/разовое задание", 
                 "Волонтерство", "Стажировка", "Оформление по ГПХ или по совместительству"]
select_checkbox(employment, driver)


# График работы
work_schedule = ["Полный день"]
for w_s in work_schedule:
    assert w_s in ["Полный день", "Сменный график", "Гибкий график", 
                   "Удаленная работа", "Вахтовый метод"]
select_checkbox(work_schedule, driver)


# Подработка
side_jobs = ["По выходным"]
for s_j in side_jobs:
    assert s_j in ["Разовое задание", "Неполный день", "От 4 часов в день", 
                   "По выходным", "По вечерам"]
select_checkbox(side_jobs, driver)


# Другие параметры
other_params = ["Меньше 10 откликов"]
other_params_dict = {
    "С адресом": "item-label_with_address", 
    "Доступные людям с инвалидностью": "item-label_accept_handicapped",
    "Без вакансий от кадровых агентств": "item-label_not_from_agency",
    "Доступные с 14 лет": "item-label_accept_kids",
    "От аккредитованных ИТ-компаний": "item-label_accredited_it",
    "Меньше 10 откликов": "item-label_low_performance",
    }

for o_p in other_params:
    assert o_p in ["С адресом", "Доступные людям с инвалидностью", "Без вакансий от кадровых агентств", 
                   "Доступные с 14 лет", "От аккредитованных ИТ-компаний", "Меньше 10 откликов"]
other_params_processed = ["advanced-search__label-" + other_params_dict[o] for o in other_params]
select_checkbox(other_params_processed, driver, type="data-qa")
```
